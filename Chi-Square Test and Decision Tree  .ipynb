# Analysis - 2 Using Chi-Square Test

```python
# Importing libraries

import pandas as pd
from scipy.stats import chi2_contingency
```

```python
# Read starbbucks data

Starbucks_Data = pd.read_csv('Starbucks satisfactory survey encode cleaned.csv', header=0)
Starbucks_Data
```

```python
# Defining the table
data_crosstab = pd.crosstab(Starbucks_Data['serviceRate'], Starbucks_Data['wifiRate'],margins = False)
print(data_crosstab)

data = data_crosstab.values.tolist()
print(data)
print(type(data))
stat, p, dof, expected = chi2_contingency(data)

# Interpret p-value
alpha = 0.05
print("p value is " + str(p))
if p <= alpha:
    print('Dependent (reject H0)')
else:
    print('Independent (H0 holds true)')
```

```python
# Defining the table
data_crosstab = pd.crosstab(Starbucks_Data['serviceRate'], Starbucks_Data['ambianceRate'],margins = False)
print(data_crosstab)

data = data_crosstab.values.tolist()
print(data)
print(type(data))
stat, p, dof, expected = chi2_contingency(data)

# Interpret p-value
alpha = 0.05
print("p value is " + str(p))
if p <= alpha:
    print('Dependent (reject H0)')
else:
    print('Independent (H0 holds true)')
```

# Building Decision Tree Model

Since the features in our dataset are categorical variables, we opted Decision Tree for building our model as Linear Regression works only on continuous variables.

```python
features = Starbucks_Data[['visitNo', 'location']]
```

```python
labels = Starbucks_Data['loyal'].to_frame()
```

```python
# Importing libraries

from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
# Splitting the dataset for train and test

X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size = 0.20, stratify = labels, random_state = 123)
```

```python
model = DecisionTreeClassifier()

# Training the model
model = model.fit(X_train, y_train)
```

```python
# Predicting the test set values
y_pred = model.predict(X_test)
```

```python
#Checking Accuracy of the model

print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
