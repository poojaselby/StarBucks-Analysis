# Data Exploration Analysis

```python
# Importing libraries

import pandas as pd
import numpy as np
from pandas import DataFrame
```

```python
# Read starbucks data

Starbucks_Data = pd.read_csv('Starbucks satisfactory survey encode cleaned.csv', header=0)
Starbucks_Data
```

```python
# Checking datatype for each variable

stb_dtype = Starbucks_Data.dtypes.to_frame()
stb_dtype.columns = ['Data_Type']
stb_dtype
```

```python
# Taking count of each categorical variable

col = Starbucks_Data.columns.values.tolist()
list_var1 = []

for i in range(1,Starbucks_Data.shape[1]):
    DataFrame(list_var1.append(Starbucks_Data[col[i]].value_counts()))
list_var1 = pd.DataFrame(list_var1)

list_var1.index.name = 'var_name1'
list_var1.reset_index(inplace=True)
list_var1
```

```python
# Reshape data by changing columns into rows based on melt function

list_var_melt = pd.melt(frame = list_var1, id_vars = 'var_name1' , var_name = 'values' , value_name = 'values_count').dropna()
list_var_melt.sort_values(['var_name1', 'values'],inplace=True, ascending=True)
list_var_melt
```

```python
# To get percentage from each categorical groups

val_sum = list_var_melt.groupby(['var_name1', 'values']).agg({'values_count': 'sum'})
val_per = val_sum.groupby(level=0).apply(lambda x:
                                                 100 * x / float(x.sum()))

val_per['values_count1'] = list_var_melt.iloc[:,2].values
var_summary = val_per.iloc[:,[1,0]]
var_summary.columns = ['value_count', 'values%']
var_summary
```

```python
# Saving result as csv file

var_summary.to_csv('starbucks_var_summary.csv', float_format='%.2f')
```
